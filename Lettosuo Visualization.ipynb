```python
plt.figure(figsize=(30,15))
plt.subplot(2,1,1)
sns.boxplot(x=letto.Place, y=letto.Flux,linewidth=4.0, data=letto)
plt.xlabel("",size=16)
plt.ylabel("Flux (g CO2/m2h)", size=16)
plt.title("Plot wise Flux distribution", size=20)
plt.show

plt.subplot(2,1,2)
sns.boxplot(x=letto.Place, y=letto.Ts5cm,linewidth=4.0,data=letto)
plt.xlabel("Plot Number",size=16)
plt.ylabel("Soil Temperature (in C)", size=16)
plt.title("Plot wise soil temperature distribution", size=20)
plt.savefig("PlotWiseFluxTemperatureDistribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x="WT", y="Flux",linewidth=4.0, data=letto)
plt.xlabel("Plot Number",size=14)
plt.ylabel("Water Table Level (in cm)", size=14)
plt.title("Plot wise water table label", size=16)
plt.savefig("WTPlotNumber.png")
plt.show()
```

```python
# Figure-level and axes-level functions
f, axes = plt.subplots(3, 2, sharey=True, figsize=(20, 15))
sns.boxplot(x="Direction", y="Flux", linewidth=4.0, data=letto, ax=axes[0,0])
sns.scatterplot(x="Ts5cm", y="Flux", hue="Treatment", data=letto, ax=axes[0,1])


sns.scatterplot(x="WT", y="Flux", hue="Treatment", data=letto, ax=axes[1,0])

sns.boxplot(x="Site", y="Flux",linewidth=4.0, data=letto, ax=axes[1,1])
sns.scatterplot(x="Site", y="Flux", hue="Treatment", data=letto, ax=axes[2,0])

sns.scatterplot(x="Distance", y="Flux", hue="Treatment", data=letto, ax=axes[2,1])
plt.savefig("MixturePlot.png")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x="Direction", y="Flux",linewidth=4.0, data=letto)
plt.xlabel("Direction",size=14)
plt.ylabel("Flux (g CO2/m2h)",size=14)
plt.title("Direction wise Flux distribution",size=16)
plt.savefig("DirectionWiseFlux", transparent=True)
plt.show()
```
